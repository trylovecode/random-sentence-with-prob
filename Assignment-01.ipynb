````python
import random
import jieba
import pandas as pd
import re
from collections import Counter
````

````python
filename = '/Users/zhangxiaonan/Desktop/train.txt'
# filename = 'train.txt'
content = pd.read_csv(filename, header=None, encoding='utf-8', delimiter="\t")
content.head()
````

````python
save=[]
result=''
pattern = re.compile("[\u4e00-\u9fa5]+?")

for line in content[0]:
    m = pattern.findall((line))
    # 将每行的中文字符进行拼接
    if m: 
        save = ''.join(str(i) for i in m)
        result+=save
print(result)
    
````

````python
# 写到另外一个txt文件里面
with open('article.txt', 'w') as f:
    for a in result:
        f.write(a)
````

````python
def cut(string): return list(jieba.cut(string))
TOKEN = []
for i, line in enumerate(open('article.txt')):
    if i > 10000: break
    TOKEN += cut(line)

words_count = Counter(TOKEN)
words_count.most_common(10)
````

````python
TOKEN = [str(t) for t in TOKEN]
# print(TOKEN)
# 两个词两个词在一起
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-2]))]
# print(TOKEN_2_GRAM[:10])
words_count_2 = Counter(TOKEN_2_GRAM)
words_count_2.most_common(10)
````

````python
# 两个词出现的概率，P(W2|W1)=P(W1+W2)同时出现 / P(W1)
def prob_2(word1, word2):
    if word1 + word2 in words_count_2: 
        return words_count_2[word1+word2] / words_count[word1]
    
    else:
        return 1 / len(TOKEN_2_GRAM)
print(prob_2('健康','保险'),prob_2('我','可以'),prob_2('家庭','保险'))
````

能够让list按照第1个元素进行排序, 但是是递减的顺序。

````python
def get_probablity(sentence):
    words = cut(sentence)
    
    sentence_pro = 1
    
    for i, word in enumerate(words[:-1]):
        next_ = words[i+1]
        
        probability = prob_2(word, next_)
        
        sentence_pro *= probability
    
    return sentence_pro
````
